## Linear weights for Run contribution
***

There is a formula developed by Pete Palmer to calculate run contribution for a player much more accurately than RBI. It's similar to the Run Expectancy and wOBA formulas that we looked at earlier in the semester, where each event in a game is assigned a weight representing its contribution to a run. The weights here are used to calculate a player's contribution above an average player's production.

$$ LWTS = (0.46)1B + (0.80)2B + (1.02)3B +(1.40)HR + (0.33)(BB+HBP) + (0.30)SB - (0.60)CS - (0.25)(AB-H)$$

Palmer claims that there is a strong correlation between a player's LWTS score and their OPS. Calculate the $R^2$, and the regression line between LWTS and OPS and plot the regression line on a scatterplot with the data. Is Palmer correct?

Recall, $$OPS = OBP + SLG$$

```python
dfTW["LWTS"] = 0.46*dfTW["1B"] + (0.80)*dfTW["2B"] + (1.02)*dfTW["3B"] +(1.40)*dfTW["HR"] + (0.33)*(dfTW["BB"]+ dfTW["HBP"]) + (0.30)*dfTW["SB"] - (0.60)*dfTW["CS"] - (0.25)*dfTW["NotHits"]
dfTW["OPS"] = dfTW["OBP"] + dfTW["SLG"]
r_2 = dfTW["LWTS"].corr(dfTW['OPS'], method='pearson')**2

print ("R^2 = ",r_2)

xmean = dfTW['LWTS'].mean()
ymean = dfTW['OPS'].mean()

std_devx = dfTW['LWTS'].std()
std_devy  = dfTW['OPS'].std()

r = dfTW['LWTS'].corr(dfTW['OPS'], method='pearson')

m = r*(std_devy/std_devx)
b = ymean - m*xmean

y = m * dfTW['LWTS'].copy() + b
x = dfTW['LWTS'].copy()
fig, ax = plt.subplots()
scatter = ax.scatter(x,dfTW['OPS'],marker='o',label="LWTS vs. OPS")
line = ax.plot(x,y, color='red',label="LWTS vs. OPS predicted")
ax.set_xlabel("LWTS", fontsize=12)
ax.set_ylabel("OPS", fontsize=12)
plt.legend()
```

Palmer was pretty accurate.

## Where do we go from here to show the effects of the ballpark?
***

The data that we used here was for individual players who play half their games at a park that is either a pitchers, hitters, or neutral park. How would we actually more definitively that the ballpark has an effect on player statistics? What data would we need that we haven't used in this assignment? How would we apply the analysis here to that data set?

We would have to tie in the ballpark the players were in to their performance at the game and take an average of how certain players (pitchers, hitters etc) played at those ballparks. We would need game by game performance for each player that also includes what ballpark they were playing in. We can aggregate player performance to a team and see how each team performs at certain ballparks and see if we can draw a correlation between ballparks and certain statistics. That way we could see how many hits, runs etc. take place in each park and see if a certain park is predisposed to giving up more hits etc.
