The primary performance metric that I used to evaluate the performance of the offense was the equivalent average statistic. The equivalent average represents a hitters productivity using the same scale as batting average. More specifically, equivalent average is the measure of offensive value per out. The EqA metric, on a basic level, is the bases a batter gets represented on the numerator and total plate appearances plus the number of times the batter has been caught stealing (we want to account for this since this is an out the batter has gained).

$$EqA = \frac{H+TB+{1.5 \cdot (BB+HBP)}+SB+SH+SF}{AB+BB+HBP+SH+SF+CS+\frac{SB}{3}}$$

- **H**: Hits 
- **TB**: Total Bases
- **SH**: Sacrifice Hits
- **SF**: Sacrifice Flyes
- **HBP**: Hit by Pitch
- **BB**: Walks
- **CS**: Caught Stealing
- **SB**: Stolen Bases

In order to find this data in Retrosheet, I had to pull out the offensive statistics of all the Rockies players who were on the BAT_TEAM_ID variable for the games they played. Most of the data I pulled was from the events table. I used the EVENT_CD attribute to match on each variable used in the equivalent average formula. The only variable I had to calculate was the total bases but I pulled each type of hit and included that in the csv file I extracted from Retrosheet so I could use pandas to sum up and create the total bases in the data frame. 

Other performance metrics I used was batting average and RBI to determine how hitters performed at Coors versus everywhere else. However, this method doesn't really take into account home field advantage. By extracting all of the offensive stats for the players, it was pretty easy to manipulate the data from the csv file in pandas.

## 1.2 Homefield Advantage

Regarding the equivalent average statistic, the variable is already weighted to be context independent of park and league effects so I had to do little work there to get rid of the homefield advantage the Rockies have. However, since I didn't do much research on the variable, I calculated two separate equivalent averages, one for the batters performance at Coors and another for the batters performance at other parks.

For statistics like the batting average and RBI, I followed a similar approach and compared the Rockies' offense at home versus at every other ball park they played at.

## 1.3 Analytics

```python
import numpy as np 
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
```

```python
local_path = 'data/col_bat_home.csv'  

dfHome= pd.read_csv(local_path)

# Inspect some of the data
dfHome.head()
```

```python
local_path1 = 'data/col_bat_away.csv'

dfAway = pd.read_csv(local_path1)

dfAway.head()
```

```python
# Equivalent Average Calculation for Coors Field Rockies Batters

dfEA_H = dfHome.groupby('Hitter').mean().reset_index()

# H+TB+{1.5 \cdot (BB+HBP)}+SB+SH+SF}/ AB+BB+HBP+SH+SF+CS+\frac{SB}{3}

dfEA_H['EqA'] = (dfEA_H['H'] + dfEA_H['TB'] + (1.5 *(dfEA_H['BB'] + dfEA_H['HBP'])) + dfEA_H['SB'] + dfEA_H['SH'] + dfEA_H['SF'])/(dfEA_H['AB'] + dfEA_H['BB'] + dfEA_H['HBP'] + dfEA_H['SH'] + dfEA_H['SF'] + dfEA_H['CS'] + (dfEA_H['SB']/3))

dfEA_H.head()
```

```python
# Equivalent Average Calculation for Rockies Batters at other fields

dfEA_A = dfAway.groupby('Hitter').mean().reset_index()

dfEA_A['EqA'] = (dfEA_A['H'] + dfEA_A['TB'] + (1.5 *(dfEA_A['BB'] + dfEA_A['HBP'])) + dfEA_A['SB'] + dfEA_A['SH'] + dfEA_A['SF'])/(dfEA_A['AB'] + dfEA_A['BB'] + dfEA_A['HBP'] + dfEA_A['SH'] + dfEA_A['SF'] + dfEA_A['CS'] + (dfEA_A['SB']/3))

dfEA_A.head()
```

- The calculations we did above gives the equivalent average for every offensive player on the Rockies at home games and at away games

```python
# Initialize figure 
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16,8))

# Plot histogram 
dfEA_H.boxplot(column="EqA", ax=ax[0]);
dfEA_A.boxplot(column ="EqA", ax=ax[1]);

ax[0].set_title("Batter EqA at Coors", fontsize=10)
ax[1].set_title("Batter EqA Away", fontsize=10)
```

- From the scatterplots above we can see that the median EqA of all the offensive players at Coors Field is higher than the median EqA of batters at away games

## 1.4 Conclusion

After calculating the average EqA for every player who has batted for the Rockies and graphing the results in the scatter plot, we can see that Rockies' offense does not underperform at home. Of course, this is assuming that the equivalent average metric is truly ball park or league independent. The median EqA for the Rockies at Coors Field is clearly higher than the median EqA for the Rockies at away games which shows us our performance is indeed better at home. There are still other variables that we aren't considering when taking into account home field advantage like the performance bump the Rockies receive by playing at Coors due to their experience at that place.

We can identify the worst performing offensive players by seeing how high the players EqA is and looking at what position they were contracted to play. There are quite a few players in the home and away dataframes who were contracted to be pitchers but they still have really low or really high EqA values due to batting in a one-off situation. We also have to consider pinch hitters and defensive players like short stops since the EqA values won't really apply to them as much as regular batters.

We can see that for away games Cole Garner has the lowest EqA, I would advise on looking for a replacement for him. For home games, Mark Ellis is one of the batters that doesn't perform well even at Coors Field.
